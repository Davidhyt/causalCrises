```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('output3.xlsx', index_col = 0)

X = df.iloc[:, 6:-1]
y = df['Outcome']
model = sm.OLS(y,sm.add_constant(X)).fit()

model.summary()
```

```python
X = df.iloc[:, 27:-1]
y = df['Crisis']
model = sm.OLS(y,sm.add_constant(X)).fit()

model.summary()
```

```python
num1 = sum(model.resid * df['Crisis'])
num1
```

```python
list1 = []
for i in range(len(list(model.resid))):
    a = list(model.resid)[i]*sum(df['Crisis'])/num1
    list1.append(a)
```

```python
df1 = df.copy()
df1['resid'] = model.resid
```

```python
df2 = pd.pivot_table(df1, values = 'Crisis', index=['Countries'],  columns=['Year'])
```

```python
df2.loc['Nicaragua',:] = df2.loc['Nicaragua',:]*2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=1)
fig, ax = plt.subplots(figsize=(18,12))
ax = sns.heatmap(df2, cmap='Reds', vmin = 0, vmax = 2)
plt.savefig('crisis_ni.pdf')
# higher resid means that higher contribution to beta 0.
```

```python
df2 = pd.pivot_table(df1, values = 'resid', index=['Countries'],  columns=['Year'])
max1 = abs(df2.max().max())
min1 = abs(df2.min().min())
a = max(max1, min1)
sns.set(font_scale=1)
fig, ax = plt.subplots(figsize=(18,12))
ax = sns.heatmap(df2, cmap='vlag', vmin = -a, vmax = a)
plt.savefig('countries.pdf')
```
